## Condicionais

No nosso dia a dia, ﾃｩ comum executarmos comportamentos diferentes para condiﾃｧﾃｵes diferentes, certo? Vamos examinar a seguinte frase:

   - **Se** chover amanhﾃ｣, leve um guarda chuva, **caso contrﾃ｡rio**, leve um ﾃｳculos de sol

De maneira anﾃ｡loga, frequentemente quando formos criar programas, vamos querer que o computador execute um dado comando apenas quando uma certa condiﾃｧﾃ｣o for verdade. Para isso, vamos usar o comando `if` e `else`: 

`````python
chuva = True

if chuva:
    print('Devo levar um guarda chuva!')
else:
    print('Devo levar um ﾃｳculos de sol!')
`````

Alﾃｩm do `if` e do `else` tambﾃｩm podemos aninhar outras condicionais entre esses comandos com o `elif`(abreviaﾃｧﾃ｣o de else if). Nﾃ｣o hﾃ｡ limites para nﾃｺmeros de `elif`s que podem ser adicionados. 
A sintaxe das condicionais vai ser a seguinte: 

``` python
if CONDIﾃﾃグ_1:
    COMANDOS_1     # executados se CONDIﾃﾃグ_1 tem valor True

elif CONDIﾃﾃグ_2:
    COMANDOS_2     # executados se CONDIﾃﾃグ_2 tem valor True

else:
    COMANDOS_2     # executados se CONDIﾃﾃグ_1 e CONDIﾃﾃグ_2 tem valor False
```

## Loops

Os loops sﾃ｣o utilizados para quando queremos que uma sﾃｩrie de comandos de cﾃｳdigos seja feita repetidamente atﾃｩ uma certa condiﾃｧﾃ｣o ou nﾃｺmero de repetiﾃｧﾃｵes ser satisfeita

### While Loops
Os loops while sﾃ｣o usados para repetir os comandos atﾃｩ uma determinada condiﾃｧﾃ｣o booleana ser satisfeita. Vale destacar que se essa condiﾃｧﾃ｣o nﾃ｣o ﾃｩ inicialmente satisfeita, o loop while nﾃ｣o roda.

`````python
#Exemplo 1
n = 1
while n <= 5:
    print(n)
    n += 1 # isso ﾃｩ o mesmo que escrever n = n + 1
`````

`````python
#Exemplo 2
n = 6
while n <= 5:
    print(n)
    n += 1
`````

Agora vamos testar uma aplicaﾃｧﾃ｣o um pouco mais robusta. Utilizando o loop while, calcularemos o fatorial de um nﾃｺmero. Lembrando que o fatorial ﾃｩ a multiplicaﾃｧﾃ｣o do nﾃｺmero pelo nﾃｺmero imediatamente anterior e assim sucessivamente atﾃｩ chegar no nﾃｺmero 1. Ex.: 3! = 3x2x1

`````python
#Fatorial
num = int(input('Escolha um nﾃｺmero! '))
multiplicador = num
fatorial = 1
while multiplicador > 1: 
    fatorial *= multiplicador
    multiplicador -= 1
print("O fatorial de %d ﾃｩ %d"%(num,fatorial))
`````

### For loops
Os loops com "for" iteram dentro de uma dada sequﾃｪncia. Normalmente, ﾃｩ usado quando queremos que uma sﾃｩrie de comandos sejam repetidos por uma quantidade de vezes especﾃｭfica. 

`````python
#Exemplo 1 - Queremos criar uma lista com os nomes de um determinado nﾃｺmero de alunos
num = int(input("Insira o nﾃｺmero de alunos: "))
lista = []
for i in range(num):
    nome = input("Insira o nome do aluno: ")
    lista.append(nome)
print(lista)
`````

Outra aplicaﾃｧﾃ｣o ﾃｩ para iterar em uma lista. 

`````python
# Vamos fazer um cﾃｳdigo que devolve os nﾃｺmeros pares de uma lista de nﾃｺmeros dados
numeros = [4,0,8,17,58,29]
for num in numeros:
    if num % 2 == 0:
        print(num)
`````

### Elementos Break e Continue
Tem algumas funﾃｧﾃｵes que podem nos ajudar a controlar o fluxo dos nossos loops. O "continue" faz com que pulemos o bloco de comando e voltemos para o comando do loop. Jﾃ｡ o "break" ﾃｩ usado para quebrar o loop e, portanto, sair dele.

`````python
#Exemplo 1: um cﾃｳdigo que printa os nﾃｺmeros impares entre 0 e 10.
for i in range(10):
    if i % 2 == 0:
        continue #se o nﾃｺmero for par, o cﾃｳdigo voltarﾃ｡ para o "for" e o nﾃｺmero nﾃ｣o serﾃ｡ printado
    print(i)
`````

`````python
#Exemplo 2: printar todos os nﾃｺmeros entre 1 e 10 que sﾃ｣o menores que 5
for i in range(10):
    if i == 5:
        break
    print(i)
`````

### Loop dentro de loop
Hﾃ｡ casos em que podemos precisar colocar vﾃ｡rios loops encadeados. 

`````python
#Exemplo 1: vamos criar um cﾃｳdigo que recebe uma lista com listas e retorna uma lista com os 
#nﾃｺmeros presentes em todas essas listas sem repetiﾃｧﾃ｣o
lista = [[1,2,3,4], [2,5,7],[15,2,20],[15,10,62]]
lista_sep = []
for i in lista:
    for j in i:
        if j not in lista_sep:
            lista_sep.append(j)
print(lista_sep)
`````

`````python
primos = []
for i in range(1,50):
    divisores = 0
    for d in range(1,i+1):
        if i % d == 0:
            divisores += 1
    if divisores == 2:
        primos.append(i)
print(primos)
`````

# Funﾃｧﾃｵes
## O que sﾃ｣o Funﾃｧﾃｵes?

Uma funﾃｧﾃ｣o ﾃｩ um bloco de cﾃｳdigo que contﾃｩm uma sequﾃｪncia de comandos que executa alguma tarefa. Sﾃ｣o utilizadas para executar determinadas tarefas que normalmente precisam ser executadas diversas vezes dentro de uma aplicaﾃｧﾃ｣o. Dessa forma, nﾃ｣o ﾃｩ necessﾃ｡rio repetir as mesmas linhas de cﾃｳdigo sempre que se quiser realizar determinada tarefa! 

## E como ﾃｩ a estrutura de uma funﾃｧﾃ｣o?

A funﾃｧﾃ｣o pode apenas manipular os valores...

```
def NOME (PARAMETROS):
  COMANDOS
```

ou retornar algum valor (usando a palavra reservada return)..

```
def NOME (PARAMETROS):
  COMANDOS
  return VALOR_DESEJADO
```

## Parﾃ｢metros opcionais e Parﾃ｢metros prﾃｩ definidos

#### Parﾃ｢metros opcionais
Algumas vezes vocﾃｪ quer usar diferentes parﾃ｢metros para diferentes propﾃｳsitos, porﾃｩm dentro da mesma funﾃｧﾃ｣o. As vezes vocﾃｪ nﾃ｣o quer usar todos os parﾃ｢metros. Existem n motivos pelos quais vocﾃｪ pode querer fazer um parﾃ｢metro ser opcional.

##### Mas como a gente faz isso?

Quando estamos escrevendo a funﾃｧﾃ｣o, declaramos o parﾃ｢metro como ```**kwargs```. Por convenﾃｧﾃ｣o, utiliza-se a palavra kwargs, mas pode-se utilizar qualquer uma, desde que coloque-se os ``**`` antes dela!
Alﾃｩm disso, ﾃｩ necessﾃ｡rio se atribuir o valor dos parﾃ｢metros opcionais ﾃ variﾃ｡veis dentro da funﾃｧﾃ｣o. A estrutura fica assim entﾃ｣o:

```` 
def nome(parametro_obrigatorio, **kwargs):
    opcional1=kwargs.get('opcional1')
    opcional2=kwargs.get('opcional2')
    ##resto da funﾃｧﾃ｣o... 
````

##### Legal, mas como eu chamo uma funﾃｧﾃ｣o com argumentos opcionais??

Posso chamar sﾃｳ com o parﾃ｢metro obrigatﾃｳrio, como uma funﾃｧﾃ｣o normal

``nome(parametro_obrigatorio)``

Posso chamar com um parﾃ｢metro opcional. Nesse caso, observe que tive que declarar que o parﾃ｢metro a mais que estava colocando na chamada da funﾃｧﾃ｣o, se refere ao parﾃ｢metro opcional1.

``nome(parametro_obrigatorio, opcional1="amendoim")``

E, por fim, posso chamar com todos os parﾃ｢netros! Sem esquecer ﾃｩ claro, de separa-los e identifica-los.

``nome(parametro_obrigatorio, opcional1="amendoim", opcional2="chuchu")``

#### Parﾃ｢metros prﾃｩ-definidos
Existem casos em que vocﾃｪ estﾃ｡ lidando com uma situaﾃｧﾃ｣o que acontece com muita frequﾃｪncia, mas nﾃ｣o pode assumir que vai acontecer sempre. Uma soluﾃｧﾃ｣o para economizar tempo nessas situaﾃｧﾃｵes, ﾃｩ prﾃｩ-definir um valor para um parﾃ｢metro da sua funﾃｧﾃ｣o. Dessa forma, o usuﾃ｡rio pode preencher o campo desse parﾃ｢metro, caso ele nﾃ｣o seja seu caso padrﾃ｣o, ou ele pode nﾃ｣o preencher nada caso se encaixe.

##### Na declaraﾃｧﾃ｣o da funﾃｧﾃ｣o

```` 
def nome(parametro_obrigatorio, parametro_padronizado = 3):
    ##resto da funﾃｧﾃ｣o... 
````
##### Na chamada da funﾃｧﾃ｣o
Posso chamar a funﾃｧﾃ｣o sem colocar nada no lugar do parametro prﾃｩ definido, e nesse caso a funﾃｧﾃ｣o irﾃ｡ utilizar o valor prﾃｩ definido
``nome(parametro_obrigatorio)``
Mas tambﾃｩm posso querer mudar o valor, nesse caso a funﾃｧﾃ｣o ﾃｩ chamada dessa forma:
``nome(parametro_obrigatorio, 5)``

### Um exemplo

`````python
def amizade(nome, mode="amigos", **kwargs):
  obs1=kwargs.get('obs1')
  obs2=kwargs.get('obs2')
  if mode == "amigos":
    print("Oi,", nome, "somos bons amigos")
    if obs1 != None:
      print(obs1)
    if obs2!= None:
      print(obs2)
  elif mode == "namorados":
    print ("oi amor, te amo, quero ter", nome, "gravado na minha alianﾃｧa")
    if obs1 != None:
      print(obs1)
    if obs2!= None:
      print(obs2)
  elif mode == "inimigos":
    print ("sai daqui,",nome, "ninguem gosta de vocﾃｪ")
    if obs1 != None:
      print(obs1)
    if obs2!= None:
      print(obs2)
`````

`````python
#Caso em que coloco os dois parﾃ｢metros opcionais, aqui ﾃｩ utilizado o valor
#prﾃｩ definido para o parﾃ｢metro 'mode'

amizade("vick", obs1="vocﾃｪ ﾃｩ cheirosa", obs2="vocﾃｪ ﾃｩ muito dedicada")
`````

`````python
#Caso sem nenhum parﾃ｢metro opcional, aqui o valor
#prﾃｩ definido para o parﾃ｢metro 'mode' ﾃｩ alterado
amizade("chuchu", "namorados")
`````

`````python
#Caso em que coloco apenas um parﾃ｢metro opcional, aqui o valor
#prﾃｩ definido para o parﾃ｢metro 'mode'ﾃｩ alterado
amizade("menino que nﾃ｣o fez nada no meu trabalho de mecﾃ｢nica", "inimigos", obs1="VOCﾃ NEM RESPONDEU MINHAS MENSAGENS NO WPP")
`````
